```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data 
```

```python
data = pd.read_csv('orders.csv')
```

```python
food_item = 'Lamb Biryani'
```

```python
# Group the data by order date and calculate the total order quantity 
```

```python
data['Year'] = pd.to_datetime(data['Order Date']).dt.year
filtered_data = data[data['Item Name'] == food_item]
grouped_data = filtered_data.groupby('Year').sum()['Quantity']
```

```python
# Plot the line chart 
```

```python
plt.plot(grouped_data.index, grouped_data.values)
plt.xlabel('Year') 
plt.ylabel('Total Order Quantity') 
plt.title('Changes in Order Quantity for Lamb Biryani') 
plt.show() 
```

```python
# Calculate the percentage change in order quantity 
```

```python
percentage_change = grouped_data.pct_change() * 100
```

```python
 # Plot the percentage change  
```

```python
plt.plot(percentage_change.index, percentage_change.values)
plt.xlabel('Order Date')
plt.ylabel('Percentage Change in Order Quantity')
plt.title('Percentage Change in Order Quantity for Lamb Biryani')
plt.show()
```
